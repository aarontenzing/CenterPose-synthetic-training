```python
import numpy as np
import plotly.graph_objects as go

# Define the coordinates
first_set = np.array([
    [ 0.28785899, -0.52800147, -3.62451426],
    [ 0.43712716, -0.77762275, -3.33017948],
    [-0.6146254,  -0.56131302, -3.19508175],
    [-0.46535722, -0.8109343,  -2.90074697],
    [ 0.69771169,  0.85962554, -2.65553854],
    [ 0.84697987,  0.61000426, -2.36120376],
    [-0.20477269,  0.82631398, -2.22610603],
    [-0.05550451,  0.57669271, -1.93177125]
], dtype=np.float32)

second_set = np.array([
    
    [ 0.25390208, -0.36795776, -3.84095579],
    [ 0.48274947, -0.96778803, -2.80884721],
    [-0.69749741, -0.24506221, -3.5585807 ],
    [-0.46865002, -0.84489248, -2.52647211],
    [ 0.69676229,  1.19670892, -3.0298134 ],
    [ 0.92560968,  0.59687865, -1.99770481],
    [-0.2546372,   1.31960447, -2.7474383 ],
    [-0.02578981,  0.7197742,  -1.71532972]
], dtype=np.float32)

# Create the 3D scatter plot
fig = go.Figure()

# Add the first set
fig.add_trace(go.Scatter3d(
    x=first_set[:, 0], 
    y=first_set[:, 1], 
    z=first_set[:, 2],
    mode='markers',
    marker=dict(size=5, color='blue'),
    name='First Set'
))

# Add the second set
fig.add_trace(go.Scatter3d(
    x=second_set[:, 0], 
    y=second_set[:, 1], 
    z=second_set[:, 2],
    mode='markers',
    marker=dict(size=5, color='red'),
    name='Second Set'
))

# Define edges to connect points from both sets
edges_first_set = [
    (0, 1), (1, 5), (5, 4), (4, 0),  # Bottom face (first_set)
    (2, 3), (3, 7), (7, 6), (6, 2),  # Top face (first_set)
    (0, 2), (1, 3), (5, 7), (4, 6)   # Vertical edges (first_set)
]

edges_second_set = [
    (0, 1), (1, 5), (5, 4), (4, 0),  # Bottom face (second_set)
    (2, 3), (3, 7), (7, 6), (6, 2),  # Top face (second_set)
    (0, 2), (1, 3), (5, 7), (4, 6)   # Vertical edges (second_set)
]

# Add edges for the first set
for edge in edges_first_set:
    point1 = first_set[edge[0]]
    point2 = first_set[edge[1]]
    
    # Add lines for edges
    fig.add_trace(go.Scatter3d(
        x=[point1[0], point2[0]],
        y=[point1[1], point2[1]],
        z=[point1[2], point2[2]],
        mode='lines',
        line=dict(color='blue', width=2)
    ))

# Add edges for the second set
for edge in edges_second_set:
    point1 = second_set[edge[0]]
    point2 = second_set[edge[1]]
    
    # Add lines for edges
    fig.add_trace(go.Scatter3d(
        x=[point1[0], point2[0]],
        y=[point1[1], point2[1]],
        z=[point1[2], point2[2]],
        mode='lines',
        line=dict(color='red', width=2)
    ))

# Set layout properties
fig.update_layout(
    title='3D Plot of World Coordinates with Edges',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show the figure directly in the browser
fig.show(renderer="browser")
```
